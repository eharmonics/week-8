```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Define the column names for the dataset
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
                'acceleration', 'model_year', 'origin', 'car_name']

# Load the data file into a Pandas DataFrame using read_csv()
df = pd.read_csv('auto-mpg.data', delim_whitespace=True, names=column_names)

# Display the first 5 rows of the DataFrame
df.head()
```

```python
# Replace '?' values in the horsepower column with NaN
df['horsepower'] = df['horsepower'].replace('?', np.nan)

# Convert the horsepower column to numeric
df['horsepower'] = pd.to_numeric(df['horsepower'])
```

```python
df['origin'] = df['origin'].map({1: 'USA', 2: 'Asia', 3: 'Europe'})
```

```python
cylinder_counts = df['cylinders'].value_counts()
plt.bar(cylinder_counts.index, cylinder_counts.values)

# Add labels and a title to the plot
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.title('Distribution of Cylinders')

# Display the plot
plt.show()
```

```python
plt.scatter(df['horsepower'], df['weight'])

# Add labels and a title to the plot
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.title('Horsepower vs. Weight')

# Display the plot
plt.show()
```

```python
plt.boxplot([df[df['origin'] == 1]['mpg'], 
             df[df['origin'] == 2]['mpg'], 
             df[df['origin'] == 3]['mpg']])
plt.xticks([1, 2, 3], ['USA', 'Asia', 'Europe'])
plt.xlabel('Origin')
plt.ylabel('Miles per Gallon')
plt.title('Distribution of MPG by Origin (Matplotlib)')

# Display the Matplotlib plot
plt.show()

# Create a boxplot of mpg by origin using Seaborn
sns.boxplot(x='origin', y='mpg', data=df)
plt.xlabel('Origin')
plt.ylabel('Miles per Gallon')
plt.title('Distribution of MPG by Origin (Seaborn)')

# Display the Seaborn plot
plt.show()
```

The resulting plots show that the median and range of mpg are highest for cars from Asia, followed by Europe and then the USA. This suggests that cars made in Asia tend to have better fuel efficiency compared to those made in Europe and the USA.
